```
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
def read_file(format_name,options,path):
    return spark.read.format(format_name).options(**options).load(path)


format_name="csv"
options={
    'header':True,
    'delimiter':',',
    'inferschema':True
}
path='dbfs:/FileStore/sample_data.csv'
```

```
df_csv=read_file('csv',options,'dbfs:/FileStore/sample_data.csv')
```

```
df_csv.display()
```

```
df=df_csv.withColumn("Currentdate",lit(current_date()))
```

```
df.display()
```

```
df1=df.withColumn('month',add_months(col('Currentdate'),-1))
```

```
df1.display()
```

```
#1st question
output1=df.filter(col("scenario")==0).withColumn('month',add_months(col('Currentdate'),-1)).display()
```

```
df2=df_csv.withColumn("Currenttimestamp",lit(current_timestamp()))
```

```
output3=df2.withColumn('current_timestamp',to_date(col('Currenttimestamp'),'yyyy-mm-dd'))
```
